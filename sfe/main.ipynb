```python
import pymc as pm
```

```python
model = pm.Model()

with model:
  lambda_ = pm.Uniform(name='lambda_', lower=0, upper=10)
  
  # likelihood
  y_obs = pm.Poisson(name='y_obs', mu=lambda_, observed=df1)
```

```python
import matplotlib.pyplot as plt

with model:
    trace = pm.sample(2000, chains = 4, tune = 1000)
    pm.plot_trace(trace)
    plt.show()
```
